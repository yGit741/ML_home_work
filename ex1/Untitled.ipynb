```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(X, y, alpha, num_iters):
    m = X.shape[0]
    n = X.shape[1]
    theta = np.zeros((n, 1))
    J_history = []

    for i in range(num_iters):
        theta = theta - (alpha/m) * X.T @ (X @ theta - y)
        J_history.append(compute_cost(X, y, theta))

    return theta, J_history

def compute_cost(X, y, theta):
    m = y.size
    J = 1/(2*m) * np.sum((X @ theta - y)**2)
    return J

# Generate sample data
m = 100
n = 2
X = np.random.rand(m, n)
y = np.random.rand(m, 1)

# Define range of alphas to test
alphas = [0.001, 0.01, 0.1, 1]

# Run gradient descent for each alpha and store results
results = []
for alpha in alphas:
    theta, J_history = gradient_descent(X, y, alpha, 100)
    results.append((alpha, J_history))

# Plot the results
fig, ax = plt.subplots()
for alpha, J_history in results:
    ax.plot(range(len(J_history)), J_history, label='alpha = {}'.format(alpha))
ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.legend()
plt.show()
```
